```python
import pandas as pd
import csv
```

```python
file_to_load = "C:/Users/moesterling/Documents/Pandas-Homework/purchase data.csv"
```

```python
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
#Player Count
total_players = len(purchase_data_df['SN'].unique())
total_players
```

```python
average_price = purchase_data_df['Price'].mean()
average_price
```

```python
unique_items = len(purchase_data_df['Item ID'].unique())
unique_items
```

```python
purchases = len(purchase_data_df['Item ID'])
purchases
```

```python
total_revenue = purchase_data_df['Price'].sum()
total_revenue
```

```python
#Purchasing Analysis (Total)
summary_statistics_df = pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":[average_price], "Number of Purchases":[purchases], "Total Revenue":[total_revenue]})
summary_statistics_df
```

```python
#Gender Demographics
purchases_unique = purchase_data_df.drop_duplicates("SN")
player_genders = purchases_unique["Gender"].value_counts()
unique_players = purchases_unique["SN"].value_counts()
total_players = len(purchases_unique["SN"])
percentage_of_players = (player_genders / total_players) * 100
gender_demographics = pd.DataFrame({"Total Players": player_genders, "Percentage of Players": percentage_of_players})
gender_demographics
```

```python
#Purchasing Analysis (Gender)
gender_group_df = purchase_data_df.groupby(['Gender'])
gender_group_one = gender_group_df['Price'].count().rename("Purchase Count")
gender_group_average_price = gender_group_df['Price'].mean().rename("Average Price")
total_price = gender_group_one * gender_group_average_price
total_price_frame = pd.DataFrame(total_price)
gender_unique = purchase_data_df.drop_duplicates("SN")
gender_uniques = gender_unique['Gender'].value_counts().rename("Purchase Count Uniques")
avg_total_purchase_per_person = pd.DataFrame(total_price / gender_uniques)
gender_group_purchase_total_frame = pd.DataFrame(gender_group_one)
result = pd.concat([gender_group_purchase_total_frame,gender_group_average_price_frame,total_price_frame,avg_total_purchase_per_person], axis=1)
result
```

```python
#Age Demographics
purchases_unique = purchase_data_df.drop_duplicates("SN")
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data_df['Total Count'] = pd.cut(purchase_data_df['Age'], bins, labels=group_names)
age_group_df = purchase_data_df.groupby(['Total Count'])
age_unique = purchase_data_df.drop_duplicates("SN")
grouped = pd.DataFrame(age_unique['Total Count'].value_counts())
percents = pd.DataFrame((grouped / purchases) * 100)
results = pd.concat([grouped,percents], axis=1)
results
```

```python
#Purchasing Analysis (Age)
age_group = purchase_data_df.groupby(['Total Count']).count()['Price'].rename("Purchase Count")
age_group_average = purchase_data_df.groupby(['Total Count']).mean()['Price'].rename("Average Price")
age_group_total_value = purchase_data_df.groupby(['Total Count']).sum()["Price"].rename("Total Purchase Value")
age_group_uniques = purchase_data_df.groupby(['Total Count'])
age_unique = purchase_data_df.drop_duplicates("SN")
age_uniques = age_unique['Total Count'].value_counts()
avg_total_purchase_per_person = (age_group_total_value / age_uniques)
age_result_df = pd.DataFrame({"Total Purchase Value": age_group_total_value, "Average Purchase Price": age_group_average, "Purchase Count": age_group, "Average Total Purchase per Person": avg_total_purchase_per_person})
age_result_df
```

```python
#Top Spenders
user_total = purchase_data_df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
user_total_count = purchase_data_df.groupby(["SN"]).count()["Price"].rename("Purchase Count")
user_total_mean = purchase_data_df.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
result_df = pd.DataFrame({"Total Purchase Value": user_total, "Average Purchase Price": user_total_mean, "Purchase Count": user_total_count})
result_df.sort_values(by=['Total Purchase Value'], ascending=False)
result_df.head()
```

```python
#Most Popular Items
popular_items_df = purchase_data_df[['Item ID', 'Item Name', 'Price']]
popular_grouped = popular_items_df.groupby(['Item ID', 'Item Name'])
popular_grouped_count = popular_grouped['Item ID'].count().rename("Purchase Count")
popular_grouped_sum = popular_items_df.groupby(['Item ID', 'Item Name']).sum()['Price'].rename("Total Purchase Value")
popular_grouped_price = popular_grouped_sum / popular_grouped_count
summary_frame = pd.DataFrame({"Purchase Count":popular_grouped_count, "Item Price": popular_grouped_price, "Total Purchase Value": popular_grouped_sum})
summary_frame.sort_values(['Purchase Count'], axis=0, ascending=False).head()
 
```
